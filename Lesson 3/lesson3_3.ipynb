```python
# BMI 計算程式

def calculate_bmi(weight, height):
    """計算 BMI"""
    bmi = weight / (height ** 2)
    return bmi

def bmi_category(bmi):
    """根據 BMI 值判斷體重分類"""
    if bmi < 18.5:
        return "體重過輕"
    elif bmi < 24:
        return "正常範圍"
    elif bmi < 27:
        return "過重"
    elif bmi < 30:
        return "輕度肥胖"
    elif bmi < 35:
        return "中度肥胖"
    else:
        return "重度肥胖"

def main():
    print("=== BMI 計算器 ===")
    try:
        # 獲取公分 (cm) 輸入
        height_cm = float(input("請輸入身高（公分）："))
        weight = float(input("請輸入體重（公斤）："))

        # 轉換為公尺 (m)
        height_m = height_cm / 100

        # 確保身高有效
        if height_m <= 0:
            print("錯誤：身高必須大於零。")
            return

        bmi = calculate_bmi(weight, height_m)
        category = bmi_category(bmi)
        
        print(f"\n您的 BMI 為：{bmi:.2f}")
        print(f"體重分類：{category}")
    except ValueError:
        print("輸入錯誤，請輸入數字。")

if __name__ == "__main__":
    main()
```
