```python
age = int (input("請輸入年齡:"))

if age >= 18:
    print("年滿18")
    print("可以考駕照了")
else:
    print("未滿18")
    print(f"再等{18-age}年,就可以考駕照了")
print("應用程式結束")
```

```python
# 曹狀判斷

age = 26

if age < 18:
    print("未年滿18")
else:
    if age <= 25:
        print("18~25之間")
    else:
        print("26以上")
```
